```python
import ipyvolume as ipv
import numpy as np
```

```python
data = np.load("numpy_emb/source_layer1.npy")
```

```python
data
```

```python
x = data[:, 3]
y = data[:, 1]
z = data[:, 2]
```

```python
fig = ipv.figure()
scatter = ipv.scatter(x, y, z, marker=".")
```

```python
ipv.show()
```

```python
scatter.geo = 'box'
```

```python
ipv.save('example.html')
```
